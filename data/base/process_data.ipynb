```python
import pandas as pd
```

```python
BIKE_PATH = "bike_sharing.csv"
RETAIL_PATH = "online_retail.csv"
```

```python
bike_df = pd.read_csv(BIKE_PATH)
bike_df.shape
```

```python
retail_df = pd.read_csv(RETAIL_PATH)
retail_df.shape
```

```python
N = 10_000

bike_10k = bike_df.head(N).reset_index(drop=True)
retail_10k = retail_df.head(N).reset_index(drop=True)

bike_10k.shape, retail_10k.shape
```

```python
bike_10k.insert(0, "row_id", range(1, len(bike_10k) + 1))
retail_10k.insert(0, "row_id", range(1, len(retail_10k) + 1))

bike_10k[["row_id"]].head(), retail_10k[["row_id"]].head()
```

```python
bike_10k.to_csv("../standardized/bike_sharing_10k.csv", index=False)
retail_10k.to_csv("../standardized/online_retail_10k.csv", index=False)

"Saved bike_sharing/bike_sharing_10k.csv and online_retail/online_retail_10k.csv"
```
